```python
def flt_check(prompt):
    try:
        value = float(input(prompt))
        return value
    except ValueError:
        print("Invalid input. Please enter a number.")
        return flt_check(prompt)

while True:
    prompt = input("Enter 1 for Simple Interest, 2 for Compound Interest, 3 for Annuity Plan or 4 to exit: ")

    if prompt == "1":
        p = flt_check("Enter the principal amount: ")
        r = flt_check("Enter the annual interest rate (in %): ")
        t = flt_check("Enter the time in years: ")
        a = p * (1 + (r / 100) * t)
        print(f"The total amount after {t} years is: {a:.2f}")
        print()
    elif prompt == "2":
        p = flt_check("Enter the principal amount: ")
        r = flt_check("Enter the annual interest rate (in %): ")
        t = flt_check("Enter the time in years: ")
        n = flt_check("Enter the number of times interest is compounded per year: ")
        #since r is in (%), divide by 100
        a = p * (1 + (r /100)/ n) ** (n * t)
        print(f"The total amount after {t} years is: {a:.2f}")
        print()
    elif prompt == "3":
        pmt = flt_check("Enter the payment amount: ")
        r = flt_check("Enter the annual interest rate (in %): ")
        t = flt_check("Enter the time in years: ")
        n = flt_check("Enter the number of payments per year: ")
        #since r is in (%), divide by 100
        a = pmt * (((1 + (r / 100) / n) ** (n * t)) - 1) / (r / 100 / n)
        print(f"The total amount after {t} years is: {a:.2f}")
        print()
    elif prompt == "4":
        print("Exiting the program.")
        break
    else: 
        print("Invalid input.") 
        print()
```
