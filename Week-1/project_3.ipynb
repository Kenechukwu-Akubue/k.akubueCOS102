```python
def flt_check(prompt):
    try:
        value = float(input(prompt))
        return value
    except ValueError:
        print("Invalid input. Please enter a number.")
        return flt_check(prompt)
print("Annuity Plan")
PMT = flt_check("Enter the payment amount: ₦")
r = flt_check("Enter the rate of annual interest (in %): ")
R = r/100
T = flt_check("Enter the time in years: ")
N = flt_check("Enter the number of times interest is compounded per year: ")
A = PMT * (((1 + R / N) ** (N * T) - 1) / (R / N))
print(f"A payment amount of ₦{PMT:.2f} at an annual interest rate of {r:.2f}%")
print(f"The annuity plan after {T} years is, if the interest is compounded {N} times a year: ₦{A:.2f}")
```
