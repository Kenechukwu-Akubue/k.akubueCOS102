```python
def flt_check(prompt):
    try:
        value = float(input(prompt))
        return value
    except ValueError:
        print("Invalid input. Please enter a number.")
        return flt_check(prompt)
print("Simple Interest")
P = flt_check("Enter the principal amount: ₦")
R = flt_check("Enter the rate of annual interest (in %): ")
T = flt_check("Enter the time in years: ")
A = P * (1 + R / 100) ** T
print(f"A principal amount of ₦{P:.2f} at an annual interest rate of {R:.2f}%")
print(f"The amount after {T} years is: ₦{A:.2f}")
```
