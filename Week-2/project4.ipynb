```python
def int_func(prompt):
    try:
        num = int(input(prompt))
        if num < 0:
            print("Value cannot be less than zero")
            return int_func(prompt)
        return num
    except ValueError:
        print("Invalid. Enter an integer")
        return int_func(prompt)

while True:
    experience = int_func("Enter employee's years of experience: ")
    age = int_func("Enter employee's age: ")
    minimum_age = experience + 18
    if age < minimum_age:
        print("Age is invalid. Age must be at least equal to the number of years of experience + 18")
        continue

    if experience > 25:
        if age >= 55:
            atr = 5_600_000
        else:
            #Not specified in question, so ATR will be mean of 5,600,000 and 4,480,000
            atr = 5_040_000 
    elif experience > 20:
        if age >= 45:
            atr = 4_480_000
        else:
            atr = 2_990_000 #Mean of 4,480,000 and 1,500,000
    elif experience > 10:
        if age >= 35:
            atr = 1_500_000
        else:
            atr = 1_025_000 #Mean of 1,500,000 and 550,000
    elif experience == 10:
        atr = 787_500 #Mean of 1,025,000 and 550,000
    else:
        if age >= 35:
            atr = 668_750 #Mean of 787,500 and 550,000
        else:
            atr = 550_000

    print(f"This employee, {age} years old, with {experience} years of experience, has an Annual Tax Revenue of ₦{atr:,}")
    break
```
