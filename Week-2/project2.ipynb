````python
import numpy as np
def val(prompt):
    try:
        x = float(input(prompt))
        n = f"{x:.4f}".rstrip('0').rstrip('.') 
        if '.' in n:
            return float(n)
        else:
            return int(n)
    except ValueError:
        print("Enter a valid number")
        val(prompt)

def decimal_strip(num):
    num = f"{num:.4f}".rstrip('0').rstrip('.')
    return num

def A_func(prompt):
    try:
        A = val(prompt)
        return A
    except ValueError:
        print("A cannot be 0")
        return A_func(prompt)

def cubic_roots(A, B, C, D):
    a = B/A
    b = C/A
    c = D/A

    p = b - a**2 / 3
    q = (2*a**3)/27 - (a*b)/3 + c

    delta = (q/2)**2 + (p/3)**3

    u = (-q/2 + (delta)**0.5)**(1/3)
    v = (-q/2 - (delta)**0.5)**(1/3)

    x1 = u + v - a/3

    omega = complex(-0.5, ((3)**0.5)/2)
    x2 = omega*u + omega**2*v - a/3
    x3 = omega**2*u + omega*v - a/3

    return x1, x2, x3

def quartic_roots(A, B, C, D, E):
    coefficients = [A, B, C, D, E]

    x1, x2, x3, x4 = np.roots(coefficients)

    return x1, x2, x3, x4

while True:
    print("Equation: Quadratic (2nd Power),  Cubic (3rd power), Quartic (4th Power)")
    operation = input("Enter 2 (for quadratic), 3 (cubic), or 4 (quartic): ")

    if operation == "2":
        print()
        print("Ax^2 + Bx + C = 0")
        A = A_func("Enter value of A: ")
        B = val("Enter value of B: ")
        C = val("Enter value of C: ")
        D = B**2 - 4*A*C
        x1 = decimal_strip((-B + D**0.5)/(2*A))
        x2 = decimal_strip((-B - D**0.5)/(2*A))
        print(f"\nThe roots of equation \n({A})x^2 + ({B})x + ({C}) = 0 \nare {x1} and {x2}")
        print()
        break
    elif operation == "3":
        print()
        print("Ax^3 + Bx^2 + Cx + D = 0")
        A = A_func("Enter value of A: ")
        B = val("Enter value of B: ")
        C = val("Enter value of C: ")
        D = val("Enter value of D: ")
        x1, x2, x3 = cubic_roots(A, B, C, D)
        x1 = decimal_strip(x1)
        x2 = decimal_strip(x2)
        x3 = decimal_strip(x3)
        print(f"\nThe roots of equation \n({A})x^3 + ({B})x^2 + ({C})x + ({D}) = 0 \nare: {x1}, {x2} and {x3}")
        print()
        break 
    elif operation == "4":
        print()
        print("Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")
        A = A_func("Enter value of A: ")
        B = val("Enter value of B: ")
        C = val("Enter value of C: ")
        D = val("Enter value of D: ")
        E = val("Enter value of E: ")
        x1, x2, x3, x4 = quartic_roots(A, B, C, D, E)
        x1 = decimal_strip(x1)
        x2 = decimal_strip(x2)
        x3 = decimal_strip(x3)
        x4 = decimal_strip(x4)
        print(f"\nThe roots of equation \n({A})x^4 + ({B})x^3 + ({C})x^2 + ({D})x + ({E}) = 0 \nare: {x1}, {x2}, {x3}, {x4}")
        print()
        break
    else:
        print("Invalid input. Please enter 2, 3, or 4.")
        print()
````
